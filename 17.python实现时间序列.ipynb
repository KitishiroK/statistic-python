```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

 # 时间序列的可视化
独取国内主要城市机场每月客流量，时间：1995-2004

```python
airport:pd.DataFrame=pd.read_csv(
    'data/Airports.csv'
)
airport.index=pd.date_range(
    freq='M',start='1995',end='2004'
)
airport.head()
```

```python
# 北京每月客流量可视化
airport['Beijing'].plot()
```

 # 时间序列平稳性检验

```python
darwin=np.genfromtxt('data/darwin.txt')
dates=pd.date_range(start='1882',periods=len(darwin),
freq='M')
Darwin=pd.Series(darwin,index=dates)
Darwin.plot()

```

```python
# ADF检验
from statsmodels.tsa.stattools import adfuller

result=adfuller(Darwin)
adf_statistic=result[0]
p_value=result[1]
print(adf_statistic)
print(p_value) # p<0.05=>序列平稳
```

```python
# 不平稳序列
print(adfuller(airport['Beijing']))
# p>0.05=> 序列不平稳
```

```python
# 不平稳序列平稳化：差分
adfuller(airport['Beijing'].diff().dropna())
# p<0.05=>平稳
```

 # 季节分解

```python
from statsmodels.tsa.seasonal import seasonal_decompose

s=seasonal_decompose(airport['Beijing'])
s.trend.plot()
s.seasonal.plot()
s.resid.plot()
```

 # ARIMA模型拟合和预测

```python
from statsmodels.tsa.stattools import acf,pacf

lags=20
# ACF图
lag_acf=acf(airport['Beijing'],nlags=lags)
plt.vlines(range(lags),[0],lag_acf,lw=5)
plt.plot(lag_acf)
plt.axhline(y=0,linestyle=':',color='b')
```

```python
# PACF图
lag_pacf=pacf(airport['Beijing'],nlags=lags)
plt.vlines(range(lags),[0],lag_pacf,lw=5)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle=':',color='b')
```

 ## 新西兰奥克兰降水数据的ARMA拟合

```python
NZ_Rainfall:pd.DataFrame=pd.read_csv(
'data/NZRainfall.csv'
)
auckland=NZ_Rainfall['Auckland'].values
dates=pd.date_range(
    start='2000',periods=len(auckland),
    freq='M'
)
auckland=pd.Series(
    auckland,index=dates
)
auckland.plot()
```

```python
from statsmodels.tsa.arima_model import ARMA
# ARMA(6,0)
A_60=ARMA(auckland,(6,0)).fit()
# ARMA(6,0)的参数估计
print(A_60.params)
```

```python
A_60.resid.plot()
```

```python
# 预测
pred_A=A_60.predict('2011-10','2012-10')
auckland['2009':].plot(label='Sample')
pred_A.plot(label='pred')
plt.legend()
```
